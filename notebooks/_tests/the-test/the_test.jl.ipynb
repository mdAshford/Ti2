#### The Louisiana Superdome has an interior volume of 125 million ft<sup>3</sup>, covered by a 440,000 ft<sup>2</sup> roof. On a particular day, the interior air pressure gave a manometer reading of 19 inches mercury. Local atmospheric pressure is 102 kPa.

<img class="" style="width:50%" src="figures/superdome-ineart-2.png" />

1. **(10) Calculate the mass of the air inside, assuming an average temperature of 20 °C.**  

  At these tempertures and pressures, air behaves as an ideal gas. Thus,
  
  \begin{equation} 
    m = \frac{PV}{RT} = \frac{P_{inside}V_{air}}{R_{air}T_{air}} 
  \end{equation} 
  
  \begin{align*}
   R_{air} &= \pu{0.287 kJ/kg.K} \\[6pt]
   P_{inside} &= P_{outside} + \Delta P_{manometer} \\ {}
  \end{align*}

1. **(10) Calculate the net force applied to the roof by the interior/exterior air. Assume the roof is flat.**

  Free body diagram anyone?

  \begin{gather}
  \begin{array}{c}
     F_{outside} \\
     \downarrow \\
     \rule[0.5ex]{3em}{0.55pt} \textit{ roof } \rule[0.5ex]{3em}{0.55pt} \notag \\
     \uparrow \\
     F_{inside} \\
  \end{array}
  \Longrightarrow 
  \begin{array}{c}
     P_{outside}A_{\it roof} \\
     \downarrow \\
     \rule[0.5ex]{3em}{0.55pt} \textit{ roof } \rule[0.5ex]{3em}{0.55pt} \\
     \uparrow \\
     P_{inside}A_{\it roof} \\
  \end{array} \\[12pt]
  F_{net} = F_{inside} + F_{outside}
  \end{gather}


```julia
module p1_2

using Unitful

# given
V_air  = 125000000u"ft^3"
T_air  = 20u"°C" |> u"K"
z_mano = 19u"inch"
P_outside = 102u"kPa"
A_roof = 440000u"ft^2"

# tables
g = 1u"gn"
ρ_Hg  = 13.6u"g/cm^3"
R_air = 0.287u"kJ/(kg*K)"

# calculate m_air; PV=mRT
P_inside = P_outside + ρ_Hg*g*z_mano |> u"kPa"
@show P_inside

m_air = P_inside * V_air / (R_air * T_air) |> u"kg"
@show m_air


# calculate F_net
## positive z is up

F_inside  =  P_inside  * A_roof   # positive is up
F_outside = -P_outside * A_roof
F = F_inside + F_outside 

@show F |> u"lbf" F |> u"kN"

end
```

#### Given 10 kg of ammonia at 0 °C, with a specific enthalpy of 1261.97 kJ/kg.   

14. **(10) Find the mass of the ammonia [kg].**   

    $$\bbox[5px,border:2px solid #461dff;]{m = \pu{10 kg}} $$   
   <br>
  
1. **(10) Find the volume of the ammonia [m<sup>3</sup>].**  
   $$  $$
   
   We have $m$, we need $v$.
  
   1. get $x(T, h)$     
   1. use $x$ to get $v(x,T)$   
   1. then V = mv
   
  Resolve state: 
  1. get $x(T, h)$
  
     \begin{gather}

  1. use $x$ to get $v(x,T)$    
  
     \begin{gather}

    It's saturated, so 
    
    \begin{equation*}

    \end{equation*}

```julia
module p14_16

using Unitful

m  = 10u"kg"
h  = 1261.97u"kJ/kg"      # given

ℎ𝑓 =  179.96u"kJ/kg"      # tables
ℎ𝑔 = 1441.94u"kJ/kg"      # tables
𝑣𝑓 = 0.0015660u"m^3/kg"   # tables
𝑣𝑔 = 0.2892u"m^3/kg"      # tables

x = (h - ℎ𝑓)/(ℎ𝑔 - ℎ𝑓); @show x
v = 𝑣𝑓 + x*(𝑣𝑔 - 𝑣𝑓); @show v

V = m * v;  @show V

end
```

```julia
## 19. (10) Determine the heat transfer, with direction (kJin/out).

module p19

using Unitful

m = 3u"kg"   # given

𝑣_100bar_600C = 0.03837u"m^3/kg"    # tables
𝑢_100bar_600C = 3241.7u"kJ/kg"      # tables
𝑣𝑓_200C = 0.0011565u"m^3/kg"        # tables
𝑣𝑔_200C = 0.1274u"m^3/kg"           # tables
𝑢𝑓_200C = 850.65u"kJ/kg"            # tables
𝑢𝑔_200C = 2595.3u"kJ/kg"            # tables

u1 = 𝑢_100bar_600C    # tables

# get x2(200 °C, v2=v1)
v1 = 𝑣_100bar_600C 
v2 = v1
x2 = (v1 - 𝑣𝑓_200C)/(𝑣𝑔_200C - 𝑣𝑓_200C)
@show x2

# get u2(x2, T2)
u2 = 𝑢𝑓_200C + x2*(𝑢𝑔_200C - 𝑢𝑓_200C)
@show u2

# heat transfer
q_out = 3u"kg" * (u1 - u2)
@show q_out

end
```

#### Steam at 280 bar, 900 °C enters a well-insulated turbine and exits at 10 kPa and 93.9% quality.  The mass flow rate is 2.34 million kg/hr, and the turbine entrance has a diameter of 63 cm.

 
23. **(10) At what temperature does the H2O exit the turbine (°C).**   
  
  The H<sub>2</sub>O exits at 10 kPa and 93.9% quality, a saturated mix.
  
  \begin{gather*}
    \bbox[5px,border:2px solid #461dff;]{ T = T_\text{sat@10kPa} = \pu{179.9 °C}} \\{}
  \end{gather*}

1. **(10) What is the turbine’s power output (GW)?**   

    Perform the First Law around the adiabatic turbine.
    
    **mass**: trivial; one inlet, one outlet
    
    **energy**:
    
    \begin{gather*}
      \dot E_{in} &-& \dot E_{out} &=& \cancelto{\rlap{0,\ SSSF}}{\frac{\rm d}{{\rm d}t}} \dot E_{sys} \\[6pt]
      \dot mh_{in} &-& \dot mh_{out} - \dot W_{out} &=& 0 \\ {}
    \end{gather*}
    
    \begin{gather}
      \dot W_{out} = \dot m \left(h_{in} - h_{out} \right) \\[6pt]
      \begin{aligned}
        h_{in} &= h\left(\pu{280 bar, 900 °C}\right) \\
        h_{out} &= h\left(\pu{10 kPa, x = 93.9\%}\right) \\
      \end{aligned}
    \end{gather}
    <br>
   
    \begin{gather*}
      \bbox[5px,border:2px solid #461dff;]{\dot W_{out} = \pu{1.068 GW}} \\{}
    \end{gather*}

1. **(10) At what velocity does the steam enter the turbine (mph)?**

    \begin{equation}
      \dot m = \rho \vec V A \\
    \end{equation}
    
    \begin{gather}
      \vec V_{in} = \frac{\dot m v_{in}}{A_{in}} \\[6pt]
      \begin{aligned}
        \dot m &= \pu{2.34*10^6 kg/hr} \\
        v_{in} &= v\left(\pu{280 bar, 900 °C}\right) \\
        A_{in} &= \frac{\pi}{4}D_{in}^2 \\
        D_{in} &= \pu{63 cm}
      \end{aligned}
    \end{gather}

    \begin{gather*}
      \bbox[5px,border:2px solid #461dff;]{\vec V_{in} = \pu{87.364 mph}}
    \end{gather*}

```julia
module p24_25

using Unitful

### 24. (10) What is the turbine’s power output (GW)?

# given
ṁ = 2.34e6u"kg/hr"   
x_out = 0.939   
D_in = 63u"cm"

# tables
h_280bar_900°C = 4298.8u"kJ/kg"
v_280bar_900°C = 0.01873u"m^3/kg"
hf_10bar  = 762.81u"kJ/kg" 
hfg_10bar = 2015.3u"kJ/kg"

# attack
h_in  = h_280bar_900°C
h_out = hf_10bar + x_out*hfg_10bar

Ẇ_out = ṁ * (h_in - h_out) |> u"GW"
@show Ẇ_out


### 25. (10) At what velocity does the steam enter the turbine (mph)?

# attack
v_in = v_280bar_900°C
A_in = π/4 * D_in^2

V⃑_in = ṁ * v_in / A_in |> u"mi/hr"
@show V⃑_in

end
```

   W_{b,out} = \int^{V_2}_{V_1}P\,{\rm d}V 
   = P\left(V_2 - V_1\right) = P\left(2V_1 - V_1\right)
   = PV_1
  \end{equation*}
  
  \begin{equation*}
   W_{b,out} = mPv_1
  \end{equation*}
  <br>

1. **(10) What is the heat transfer for this process (kJ)?**    
  First Law:
  \begin{gather*}
    E_{in} &-& E_{out} &=& \Delta E_{sys} \\
    Q_{in} &-& W_{b,out} &=& U_2 - U_1 \\
    Q_{in} &-& P\left(V_2 - V_1 \right) &=& U_2 - U_1 
  \end{gather*}
  
  \begin{equation*}
    Q_{in} 
     = U_2 - U_1 + P\left(V_2 - V_1 \right) 
     = \left(U_2 + PV_2 \right) - \left(U_1 + PV_1 \right) 
     = H_2 - H_1 
  \end{equation*}
  
  \begin{equation*}
   Q_{in} = m\left( h_2 - h_1 \right)
  \end{equation*}
