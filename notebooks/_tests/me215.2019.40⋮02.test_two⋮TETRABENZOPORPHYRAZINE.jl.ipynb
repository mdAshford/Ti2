    <td style=""></td>
    <td colspan=2 class="">
      <p>
To find heat transfer, we employ the energy equation. The assumed direction of heat transfer is arbitrary. We have no indication of kinetic or potential energy implications, so we neglect both $\left( \Delta\ce{KE} \cong \Delta\ce{PE} \cong 0\right)$. We proceed as follows:  


\begin{gather*}
  E_{in} &-& E_{out} &=& \Delta E_{sys} \\
  Q_{in} &-& W_{out} &=& U_{final} - U_{initial} \\
  && &=& m\left( u_{final} - u_{initial}\right) 
\end{gather*}

\begin{equation*}
  Q_{in} = W_{b,out} + m\left( u_\text{500°C,500kPa} - u_\text{f@500kPa}\right) \\
\end{equation*}
        
We've already figured out the boundary work, so we could be done here. However, had work not been explicitly requested, you could've proceeded as such: 

\begin{gather*}
   E_{in} &-& E_{out} &=& \Delta E_{sys} \\[6pt]
   Q_{in} &-& W_{b,out}  &=&  U_{final} - U_{initial} \\[6pt]
   Q_{in} &-& \displaystyle \int^{\it final}_{\it initial} P\, {\rm d}V &=&  U_{final} - U_{initial} \\[6pt] 
   Q_{in} &-& \displaystyle  P\left(V_{final} - V_{initial} \right) &=&  U_{final} - U_{initial} \\ {}
\end{gather*}

\begin{array}{c}
  Q_{in} &=& U_{final} - U_{initial} \displaystyle + P\left(V_{final} - V_{initial} \right) 
    &=& \left( U_{final} + PV_{final} \right) &-& \left( U_{initial} + PV_{initial} \right) \\
    &&&=& H_{final} &-& H_{initial} \\{}
\end{array}

\begin{equation*}
  Q_{in} = m\left(h_{final} - h_{initial}\right) = m\left( h_\text{500°C,500kPa} - h_\text{f@500kPa}\right)
\end{equation*} 

\begin{equation*}
  Q_{in} = m\left(h_{final} - h_{initial}\right) = m\left( h_\text{500°C,500kPa} - h_\text{f@500kPa}\right)
\end{equation*}    
      </p>
    </td>

```julia
module p20x22

using Unitful
       
m = 0.5u"kg"   # given
P = 500u"kPa"  # given

# ① 500 kPa, satuarated liquid
vf_500kPa = 0.0010926u"m^3/kg"   # table
uf_500kPa = 639.68u"kJ/kg"       # table
hf_500kPa = 640.23u"kJ/kg"       # table

# ② 500kPa, 500°C
v_500°C_500kPa = 0.7109u"m^3/kg" # table
h_500°C_500kPa = 3483.9u"kJ/kg"  # table
u_500°C_500kPa = 3128.4u"kJ/kg"  # table


## 20. Find the final volume of the H₂O.
v_final = v_500°C_500kPa
V_final = m * v_final
@show V_final


## 21. BONUS: Determine the work done by the H2O.
v_initial = vf_500kPa
W_out = m * P * (v_final - v_initial) |> u"kJ"
@show W_out 


## 22. BONUS: Determine the heat transfer.
u_initial = uf_500kPa
u_final   = u_500°C_500kPa

h_initial = hf_500kPa
h_final   = h_500°C_500kPa


### using already calculated W_out
Q_in = W_out + m*(u_final - u_initial) |> u"kJ"


### not using already calculated W_out
Q_in_by_enthalpy = m*(h_final - h_initial) |> u"kJ"


## show your talents
println(
"""

Heat transfer, calulated two ways:
      using W_out: Q_in = $(Q_in)
  not using W_out: Q_in = $(Q_in_by_enthalpy)
"""
)

end
```
