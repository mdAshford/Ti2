#### mass
    
Don't fall for the okey-doke!! This is a closed system, so $\Delta m = 0,$ and "**final**" mass is meaningless. 

\begin{equation} 
\begin{gathered}
 \displaystyle m = \frac{V}{v},  \\
  v = {v_f}_{\rm200kPa} + x_1 \left( {v_g}_{\rm200kPa} - {v_f}_{\rm200kPa} \right)
\end{gathered}
\end{equation}



#### work and heat transfer

Apply the energy equation: 

\begin{array}{c} 
 E_{in} &-& E_{out} &=& \Delta E_{sys} \\
 Q_{in} &-& W_{out} &=& U_2 - U_1 \\ 
\end{array}
    
The whole process is isochoric, so $W_b = 0.$ Moreover, no other forms of work are mentioned. What remains is
    
\begin{gather}
 Q_{in} = m\left(u_2 - u_1\right) \tag{6} \\[6pt]
\end{gather}
  
  
#### resolve ② 

We already have $m$, and we can get $u_1$relatively easily. We resolve ② by finding where $v_g = v_2 = v_1$. This one requires interpolation.
```julia
#= Julia pro tip

modules have their own variable space. variables are loccked into the module, 
so we can reuse variable names without fear of one problem's variables from 
bleeding into another's. debugging bleed-through is a nightmare that makes one 
hate every atom in the cosmos. 

Prefix a variable name with its module name to access it:
      good:  In [1]: TwoFive.V  
    sinful:  In [1]: V    

See below for example     
                                                                            =#

module TwoFive

using Unitful   # new module, ∴ new package declarations

# given and known states
V  = 350u"m^3"               # given  
x1 = 0.25                    # given    
   
vf = 1.0605e-3u"m^3/kg"      # tables @200kPa   
vg = 0.8857u"m^3/kg"         # tables @200kPa

uf1 = 504.49u"kJ/kg"         # tables @200kPa
ug1 = 2529.5u"kJ/kg"         # tables @200kPa
u1 = uf1 + x1 * (ug1 - uf1) 

# resolve ② via interpolation
vg_up = 0.2404u"m^3/kg";   T_up = 170.4u"°C";   ug_up = 2576.8u"kJ/kg"
v = vf + x1 * (vg - vf)
vg_dn = 0.2150u"m^3/kg";   T_dn = 175.4u"°C";   ug_dn = 2580.5u"kJ/kg"


#= interpfrac: known side of the interpolation eqn:

              (  v   - vg_dn)     ( T2   -  T_dn)     ( ug2  - ug_dn)     
interpfrac = ───────────────── = ───────────────── = ───────────────── 
              (vg_up - vg_dn)     (T_up  -  T_dn)     (ug_up - ug_dn) 
=#

interpfrac = (v - vg_dn) / (vg_up - vg_dn) 
                                   
T2  = ( T_dn  + (T_up  -  T_dn) * interpfrac ) |> u"°C" 
ug2 = ( ug_dn + (ug_up - ug_dn) * interpfrac ) 
 
m  = V / v
Q_in = m * (ug2 - u1) |> u"kJ" 


@show m T2 Q_in  

end
```

<h4 id="Attack" style="color: #18a2b8;">Attack</h4>
  
1. set up energy equation
1. pull out $W_b$ and resolve
1. resolve the rest
    




Start the enery equation like we do every single time. 
    
\begin{array}{c} 
 E_{in} &-& E_{out} &=& \Delta E_{sys} \\
 Q_{in} &-& W_{b,out} &=& U_2 - U_1 \\  
\end{array}

\begin{equation} 
 Q_{in} = U_2 - U_1 + W_{b,out} \label{Qin}
\end{equation}

Expressing boundary work for an isobaric process is straightforward:
    
\begin{gather} 
 W_{b,out} = \int^2_1 P\,{\rm d}V = P \left( V_2 - V_1 \right) = mP \left( v_2 - v_1 \right) \label{boundary} \\{} \notag
\end{gather}

 
<br><br>


#### you have options for calculating the heat transfer. 
  
1. You can use the boundary work you just calculated in (\ref{boundary}). From (\ref{Qin}):

    \begin{align} 
 Q_{in} &= U_2 - U_1 + W_{b,out} \notag \\[9pt]  
   Q_{in} &= m \left( u_2 - u_1 \right) + W_{b,out} \\[9pt] 
    & \phantom{=\left( U_2 + PV_2 \right) - \left( U_1 + PV_1 \right)} \notag {}
\end{align}

1. Or apply a little algebra and solve for heat transfer using enthalpy. Expanding (\ref{Qin}):
    
    \begin{align} 
 Q_{in} &= U_2 - U_1 + W_{b,out} \notag \\[6pt]
 &= U_2 - U_1 + P \left( V_2 - V_1 \right) \notag \\[6pt]
    &= \left( U_2 + PV_2 \right) - \left( U_1 + PV_1 \right) \notag \\[6pt]
   &= H_2 - H_1 \notag \\[9pt]
   Q_{in} &= m \left( h_2 - h_1 \right) 
\end{align}


<br><br>
    


#### resolve both states, do the math, and pat yourself on the back.
    
①$\begin{array}{c|c}
 \left\{\begin{aligned}
   P_1 &= \text{5 bar} \\
   T_1 &= \text{200 °C}
 \end{aligned}\right. & \begin{aligned}
   v_1 &= 0.4249~{\rm m^3/kg} \\
   u_1 &= 2642.9~\text{kJ/kg} \\
   h_1 &= 2855.4~\text{kJ/kg}  
 \end{aligned}
 \end{array}$  
    
②$\begin{array}{c|c}
 \left\{\begin{aligned}
   P_2 &= \text{5 bar} \\
   x_2 &= \rlap{10\%}{\hphantom{\text{200 °C}}}
 \end{aligned}\right. 
 & \begin{aligned}
 {T_2} & = {151.9~{\rm °C}} \\
   {v_{f_2}} &= 0.0010926~{\rm m^3/kg} & v_{g_2} &= 0.3749~{\rm m^3/kg} \\
   u_{f_2} &= 639.68~\text{kJ/kg} & u_{g_2} &= 2561.2~{\rm kJ/kg} \\
   h_{f_2} &= 640.23~\text{kJ/kg} & u_{g_2} &= 2748.7~{\rm kJ/kg} 
 \end{aligned}
 \end{array}$

```julia
module SixNine  

using Unitful

# given
x2 = 0.1
P  = 5u"bar"
m  = 0.6u"kg"

# tables
v1 = 0.4249u"m^3/kg";   vf2 = 0.0010926u"m^3/kg";   vg2 = 0.3749u"m^3/kg"
u1 = 2642.9u"kJ/kg";    uf2 = 639.68u"kJ/kg";       ug2 = 2561.2u"kJ/kg"
h1 = 2855.4u"kJ/kg";    hf2 = 640.23u"kJ/kg";       hg2 = 2748.7u"kJ/kg"

# attack
v2 = vf2 + x2*(vg2 - vf2)
u2 = uf2 + x2*(ug2 - uf2)
h2 = hf2 + x2*(hg2 - hf2)

Wb_out =  P * m * (v2 - v1) 
Wb_in  = -Wb_out |> u"kJ" 

Q_out_byU = -(m * (u2 - u1) + Wb_out) |> u"kJ" 
Q_out_byH = -(m * (h2 - h1))          |> u"kJ" 

@show Wb_in Q_out_byU Q_out_byH

end;
```
